```python
import requests

url = "https://dods.ndbc.noaa.gov/thredds/fileServer/oceansites/DATA/CCE2/OS_CCE2_01_D_AQUADOPP.nc"

r = requests.get(url, verify=False)

with open("OS_CCE2_01_D_AQUADOPP.nc", "wb") as f:
    f.write(r.content)
```

```python
import xarray as xr

ds = xr.open_dataset("OS_CCE2_01_D_AQUADOPP.nc")

df = ds.to_dataframe().reset_index()
df.to_csv("output.csv", index=False)

print("Saved to output.csv")
```
